```python
# Carregar bibliotecas
import numpy as np # biblioteca para manipular matrizes
import pandas as pd # biblioteca para manipular e tratar os dados

#Gabriel Conforto
```

```python
 #import pandas as pd

# URL of the CSV file
url = "https://raw.githubusercontent.com/C0D4IN/CP2/refs/heads/main/phishing_featuring_cp2.csv"

# Load the CSV file into a DataFrame
df = pd.read_csv(url)

# Display the first few rows of the DataFrame to verify
df.head(10)
#icaro
```

```python
# Fabio

# Função para extrair todas as features de uma URL
def extract_features(url):
    domain = url.split('/')[2] if '//' in url else url.split('/')[0]
    directory = '/' + '/'.join(url.split('/')[3:]).split('?')[0] if '/' in url else ''
    file = directory.split('/')[-1] if '.' in directory else ''
    params = url.split('?')[1] if '?' in url else ''

    return {
        'qty_dot_url': url.count('.'),# Robert
        'qty_hyphen_url': url.count('-'),# Ícaro
        'qty_slash_url': url.count('/'),# Mateus Rocha
        'qty_equal_url': url.count('='), # Robert
        'qty_exclamation_url': url.count('!'), # Mateus Rocha
        'qty_equal_domain': domain.count('='),# Robert
        'qty_exclamation_domain': domain.count('!'),# Ícaro
        'qty_hyphen_domain': domain.count('-'), # Higor Moura
        'qty_dot_domain': domain.count('.'), # Gabriel Oliveira
        'qty_slash_domain': domain.count('/'), # Higor Moura
        'qty_dot_directory': directory.count('.'),# Gabriel Conforto
        'qty_underscore_directory': directory.count(''),# Gabriel Conforto
        'qtyslash_directory': directory.count('/'),# Gabriel Conforto
        'directory_length': len(directory),# Mateus Rocha
        'qty_slash_file': file.count('/'), # Fábio
        'qty_dot_file': file.count('.'), # Higor Moura
        'file_length': len(file),# Gabriel Conforto
        'qtyslash_file': file.count('/'),# Fábio
        'qty_dot_params': params.count('.'),# Robert
        'qty_hyphen_params': params.count('-'),# Gabriel Conforto
        'qty_underscore_params': params.count(''),# Gabriel Conforto
        'qty_slash_params': params.count('/') # Gabriel Conforto
    }

# Aplicando a função em cada linha da coluna URL e expandindo o resultado em novas colunas
df = df.join(df['url'].apply(lambda x: pd.Series(extract_features(x))))

df.head(5)
# Fábio
```
